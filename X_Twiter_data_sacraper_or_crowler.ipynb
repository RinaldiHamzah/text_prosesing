Cleaning Teks

```python
import re

def clean_text(text):
    text = str(text).lower() #input berupa string
    text = re.sub(r'http\S+|www\S+|https\S+', '', text) # Hapus URL/link
    text = re.sub(r'@\w+', '', text)  # Hapus mention @username
    text = re.sub(r'#', '', text)  #ormas → ormas
    text = re.sub(r'\d+', '', text) # Hapus angka
    text = re.sub(r'[^a-z\s]', ' ', text) # Hapus tanda baca dan karakter non-huruf
    text = re.sub(r'\brt\b', '', text) # Hapus kata "rt" (retweet)
    text = re.sub(r'\s+', ' ', text).strip()  # Hapus spasi berlebih
    return text

data['clean_text'] = data['full_text'].apply(clean_text)

print(data[['full_text', 'clean_text']].head())
```

Stemming Teks

```python
# ==== Fungsi bantu ====
def stem_word(word):
    prefixes = ['meng', 'meny', 'men', 'mem', 'me',
                'peng', 'peny', 'pen', 'pem', 'di', 'ke', 'se',
                'ber', 'ter', 'per']
    suffixes = ['kan', 'an',  'lah', 'nya', 'ku', 'mu']

    for p in prefixes:
        if word.startswith(p) and len(word) > len(p) + 2:
            word = word[len(p):]
            break

    for s in suffixes:
        if word.endswith(s) and len(word) > len(s) + 2:
            word = word[:-len(s)]
            break

    return word
```

```python
def apply_stemming(teks):
    words = teks.split()                    # pecah teks jadi kata
    stems = [stem_word(w) for w in words]   # stemming per kata
    return ' '.join(stems)                  # gabung kembali jadi kalimat

# Terapkan ke kolom clean_text
data['steam_text'] = data['clean_text'].apply(apply_stemming)

print(data[['clean_text', 'steam_text']].head())
```

```python

# Buat Stemmer Bahasa Indonesia dari Sastrawi ===
factory = StemmerFactory()
stemmer_id = factory.create_stemmer()

#Fungsi Lemmatizer Manual ===
def manual_lemmatizer_id(word):
    manual_dict = {
        'nu': 'nahdatul ulama',
        'ormas': 'organisasi masyarakat',
        'memakan': 'makan',
        'menulis': 'tulis',
        'tertidur': 'tidur',
        'bermain': 'main',
        'dilakukan': 'laku',
        'membaca': 'baca',
        'menyukai': 'suka'
    }

    if word in manual_dict:
        return manual_dict[word]

    prefixes = ['meng', 'meny', 'men', 'mem', 'me', 'ber', 'ter', 'se', 'ke', 'di', 'per']
    suffixes = ['kan', 'an', 'i', 'lah', 'nya', 'ku', 'mu']

    for p in prefixes:
        if word.startswith(p) and len(word) > len(p) + 2:
            word = word[len(p):]
            break

    for s in suffixes:
        if word.endswith(s) and len(word) > len(s) + 2:
            word = word[:-len(s)]
            break

    return word

#Fungsi Kombinasi Lemmatization ===
def combined_lemmatizer_id(text):

    text = str(text)
    words = text.split()
    result = []

    for w in words:
        manual_lemma = manual_lemmatizer_id(w)
        if manual_lemma != w:  # jika manual berhasil mengubah kata
            result.append(manual_lemma)
        else:
            auto_lemma = stemmer_id.stem(w)
            result.append(auto_lemma)

    return ' '.join(result)

#Terapkan ke kolom steaming_text ===
data['lemmatizer_text'] = data['steaming_text'].apply(combined_lemmatizer_id)

#Lihat hasil gabungan ===
print(data[['steaming_text', 'lemmatizer_text']].head())
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Gunakan model multilingual (Indonesia + English)
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Buat pipeline sentiment analysis
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
def get_sentiment(text):
    try:
        result = sentiment_pipeline(str(text))[0]
        label = result['label']

        # Model ini memberi label bintang (1–5), jadi kita ubah jadi positif/negatif
        if '1' in label or '2' in label:
            return 'negatif'
        elif '4' in label or '5' in label:
            return 'positif'
        else:
            return 'netral'
    except:
        return 'netral'

# Terapkan ke kolom lemma_bilingual
data['label_text'] = data['tokens_text'].apply(get_sentiment)
```

```python
# Convert a string-formatted list to a real Python list
import ast

def convert_text_list(texts):
    # Safely evaluate the string to a Python object (e.g., "[1, 2, 3]" → [1, 2, 3])
    texts = ast.literal_eval(texts)
    return [text for text in texts]

# Apply the conversion to the 'tweet_clean' column to create a new column 'tekslist'
data["list_text"] = data["tokens_text"].apply(convert_text_list)

# Print one example converted list
print(data["list_text"][0])

# Print the type of that element to confirm it's now a Python list
print("\ntype :", type(data["list_text"][5]))
```

Perhitungan TF

```python
def calc_TF(document):
    # Step 1: Count term occurrences
    tf_dict = {}
    for term in document:
        if term in tf_dict:
            tf_dict[term] += 1
        else:
            tf_dict[term] = 1

    # Step 2: Normalize by the total number of terms (TF calculation)
    for term in tf_dict:
        tf_dict[term] = tf_dict[term] / len(document)

    return tf_dict

# Apply TF calculation to each list of tokens in 'tekslist'
data["tf_text"] = data["list_text"].apply(calc_TF)

# Display the first few TF dictionaries
data["tf_text"].head()
```

```python
#Displaying a Document’s TF Values
#Prints all terms and their Term Frequency (TF) values for a specific document (here, document index = 1)
index = 1

print("%20s" % "term", "\t", "tf\n")
for key in data["tf_text"][index]:
    print('%20s' % key, "\t", data["tf_text"][index][key])
```

Perhitungan IDF

```python
#Calculating Document Frequency (DF)
def calc_DF(tfDict):
    count_DF = {}
    for document in tfDict:
        for term in document:
            if term in count_DF:
                count_DF[term] += 1
            else:
                count_DF[term] = 1
    return count_DF

DF = calc_DF(data["tf_text"])
```

```python
#Calculating Inverse Document Frequency (IDF)
import numpy as np
n_document = len(data)

def calc_IDF(n_document, DF):
    IDF_dict = {}
    for term in DF:
        IDF_dict[term] = np.log10(n_document / (DF[term] + 1))
    return IDF_dict

IDF = calc_IDF(n_document, DF)
```

```python
selected_terms = ['ormas', 'nu', 'hukum']

idf_selected = {term: IDF.get(term, 0) for term in selected_terms}
print("Nilai IDF untuk kata ['ormas', 'nu', 'hukum']:")
print(idf_selected)
```

```python
#Calculating TF-IDF
#Purpose:
#Computes the TF-IDF value for each term in each document by multiplying TF and IDF.
def calc_TF_IDF(TF, IDF):
    TF_IDF_dict = {}
    for key in TF:
        if key in IDF:  # add safety check
            TF_IDF_dict[key] = TF[key] * IDF[key]
    return TF_IDF_dict

# Apply properly
data["tf-idf_text"] = data["tf_text"].apply(lambda x: calc_TF_IDF(x, IDF))
```

Matrik TF-IDF

```python
# === Membuat matriks TF-IDF ===
# Urutkan kata berdasarkan Document Frequency (descending)
sorted_DF = sorted(DF.items(), key=lambda kv: kv[1], reverse=True)[:30]
unique_terms = [term for term, _ in sorted_DF]

# Fungsi untuk mengonversi TF-IDF dict ke vektor
def calc_TF_IDF_Vec(tfidf_dict):
    vector = [0.0] * len(unique_terms)
    for i, term in enumerate(unique_terms):
        vector[i] = tfidf_dict.get(term, 0.0)
    return vector

# Terapkan fungsi ke setiap dokumen
data["tf_idf_vector"] = data["tf-idf_text"].apply(calc_TF_IDF_Vec)

# Cek hasil
print("Tampilan baris pertama dari matriks tf_idf_vector:\n")
print(data["tf_idf_vector"].iloc[1])
print("\nUkuran vektor (jumlah fitur):", len(data["tf_idf_vector"].iloc[0]))
```

Menampilkan ke dalam list

```python
# === Menampilkan top 30 term berdasarkan total TF-IDF ===

# Konversi kolom TF-IDF vector menjadi array 2D
TF_IDF_Vec_List = np.array(data["tf_idf_vector"].to_list())

# Hitung jumlah TF-IDF setiap term di seluruh dokumen
term_sums = TF_IDF_Vec_List.sum(axis=0)

# Buat DataFrame untuk peringkat term
ranking = pd.DataFrame({
    'term': unique_terms,
    'rank': term_sums
})

# Urutkan dari nilai tertinggi ke terendah
ranking = ranking.sort_values(by='rank', ascending=False).reset_index(drop=True)

# Tampilkan top 30 term
print("Top 30 terms berdasarkan total TF-IDF:")
print(ranking.head(100))
```
